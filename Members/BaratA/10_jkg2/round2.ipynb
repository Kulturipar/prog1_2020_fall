## ValentinyiA

```python
def ValentinyiA_solution(list_of_numbers, number):
    l = []
    for i in list_of_numbers:
        if (i/number) % 2 == 0:
            l.append(i)
    if not l:
        return 0
    else:
        return max(l)
```

```python
ValentinyiA.evaluate(ValentinyiA_solution)
```

## KeglV

```python
def KeglV_solution(input):
    a = sorted(input)
    if a[-1] * a[-2] > a[0] * a[1]:
        return a[-1] * a[-2]
    else:
        return a[0] * a[1]
```

```python
KeglV.evaluate(KeglV_solution)
```

## TompaM

```python
def TompaM_solution(list_of_numbers, num):
    lst = []
    for i in list_of_numbers:
        if (i/num) % 2 == 0:
            lst.append(i)
    if not lst:
        return 0
    else:
        return max(lst)
```

```python
TompaM.evaluate(TompaM_solution)
```

## VelkeyA

```python
def VelkeyA_solution(lista):
    a = sorted(lista)
    b = []
    for i in range(len(a)):
        x = a[i]*2
        for j in range(len(a)):
            if a[j] == x:
                b.append(x/2)
    if not b:
        return 0
    else:
        return b[0]
```

```python
VelkeyA.evaluate(VelkeyA_solution)
```

## PapS

```python
def PapS_solution(list_of_words):
    max_a = 0
    max_a_index = -1
    for i,j in enumerate(list_of_words):
        a_num = j.lower().count('a')
        if a_num > max_a:
            max_a = a_num
            max_a_index = i
    return max_a_index        
```

```python
PapS.evaluate(PapS_solution)
```
